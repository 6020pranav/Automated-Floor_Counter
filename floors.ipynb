```python
from ultralytics import YOLO
from IPython.display import Image, display
from PIL import Image as PILImage
import numpy as np
import cv2

# Load the trained YOLO model
model = YOLO('/content/runs/detect/train4/weights/best.pt')  # Adjust the model path if needed

# List of custom input images
custom_image_paths = [
    '/content/car_logo-1/train/images/images-10-_jpeg.rf.35e3cf465a3ed9bce3ccd54de3325cca.jpg'
]

for image_path in custom_image_paths:
    # Load the image
    img = PILImage.open(image_path)

    # Convert image to a format suitable for OpenCV
    img_cv = np.array(img)

    # Perform prediction with a lower confidence threshold
    results = model.predict(img_cv, conf=0.1)

    # Print out the results for debugging
    print(f"Predictions for {image_path}: {results}")

    # Draw bounding boxes on the image
    for result in results[0].boxes:
        x1, y1, x2, y2 = map(int, result.xyxy[0])
        conf = result.conf.item()
        class_id = int(result.cls)
        label = f'{model.names[class_id]} {conf:.2f}'
        cv2.rectangle(img_cv, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(img_cv, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    # Convert image back to PIL format
    img_result = PILImage.fromarray(img_cv)

    # Save the image to display in the notebook
    img_result.save('/content/temp_result.jpg')

    # Display the image with bounding boxes
    display(Image(filename='/content/temp_result.jpg', height=600))
    print("\n")
```

```python
from ultralytics import YOLO
from IPython.display import Image, display
from PIL import Image as PILImage
import numpy as np
import cv2

# Load the trained YOLO model
model = YOLO('/content/runs/detect/train4/weights/best.pt')  # Adjust the model path if needed

# List of custom input images
custom_image_paths = [
    '/content/37845905.jpg'
]

for image_path in custom_image_paths:
    # Load the image
    img = PILImage.open(image_path)

    # Convert image to a format suitable for OpenCV
    img_cv = np.array(img)

    # Perform prediction with a lower confidence threshold
    results = model.predict(img_cv, conf=0.1)

    # Print out the results for debugging
    print(f"Predictions for {image_path}: {results}")

    # Draw bounding boxes on the image for predictions with confidence above 0.80
    for result in results[0].boxes:
        if result.conf > 0.20:
            x1, y1, x2, y2 = map(int, result.xyxy[0])
            conf = result.conf.item()
            class_id = int(result.cls)
            label = f'{model.names[class_id]} {conf:.2f}'
            cv2.rectangle(img_cv, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(img_cv, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    # Convert image back to PIL format
    img_result = PILImage.fromarray(img_cv)

    # Save the image to display in the notebook
    img_result.save('/content/temp_result.jpg')

    # Display the image with bounding boxes
    display(Image(filename='/content/temp_result.jpg', height=600))
    print("\n")
```
